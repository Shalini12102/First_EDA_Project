# **Step 5:Outlier Identification and Visualization**

## 1. Using Statistical Methods (Z-Score and IQR)

```python
for column in app_data.select_dtypes(include=['int64','float64']).columns:
  # Z-Score Method
  app_data['Z_Score'] = (app_data[column] - app_data[column].mean()) / app_data[column].std()
  outliers_z = app_data[(app_data['Z_Score'] > 3) | (app_data['Z_Score'] < -3)]

  # IQR Method
  Q1 = app_data[column].quantile(0.25)
  Q3 = app_data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers_iqr = app_data[(app_data[column] < lower_bound) | (app_data[column] > upper_bound)]

# Output results
print("Outliers identified by Z-Score:")
print(outliers_z)

print("Outliers identified by IQR:")
print(outliers_iqr)
```

# **Step 7:Top 10 Correlations Analysis (Segmented by Target)**

## Step 1: Segment Data and Calculate Correlations

```python
# 1 = clients with payment difficulties
# 0 = clients without payment difficulties
# Define the target column
target_column = 'TARGET'

# Dictionary to store top 10 correlations for each segment
top_correlations = {}

# Loop through each segment: 1 (clients with payment difficulties) and 0 (without payment difficulties)
for target_value in [0, 1]:
    # Segment the data
    segment = app_data[app_data[target_column] == target_value]
    # Calculate the correlation matrix for numeric data in this segment
    correlation_matrix = segment.select_dtypes(include=['number']).corr()

    # Unstack the matrix to get pairs in a long format
    correlations = correlation_matrix.unstack()

    # Drop self-correlations (correlation of a variable with itself)
    correlations = correlations[correlations.index.get_level_values(0) != correlations.index.get_level_values(1)]

    # Sort by absolute correlation values and drop duplicate pairs
    sorted_correlations = correlations.abs().sort_values(ascending=False).drop_duplicates()

    # Get the top 10 correlations for the segment
    top_correlations[target_value] = sorted_correlations.head(10)

    # Display the top 10 correlations for each target segment
    print(f"\nTop 10 Correlations for {target_column} = {target_value} (Clients {'with' if target_value == 1 else 'without'} Payment Difficulties):\n",
          top_correlations[target_value])
```

## Step 2: Visualize the Top Correlations

```python
# Get unique values in the target column
target_val = app_data[target_column].unique()

for value in target_val:
    # Create a DataFrame from the top correlations
    top_corr_df = top_correlations[value].reset_index()
    top_corr_df.columns = ['Variable 1', 'Variable 2', 'Correlation']


    # Bar Plot of Top Correlations
    sns.barplot(data=top_corr_df, x='Correlation', y='Variable 1', hue='Variable 2', dodge=False, palette='coolwarm')
    plt.title(f'Top 10 Correlations for {target_value} = {value}')
    plt.xlabel('Correlation')
    plt.ylabel('Variable Pair')
    plt.show()
```
